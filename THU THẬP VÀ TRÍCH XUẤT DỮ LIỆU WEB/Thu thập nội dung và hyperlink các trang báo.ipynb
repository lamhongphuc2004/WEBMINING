```python
#2.5b. Bóc tách dữ liệu nhiều trang
import requests
from bs4 import BeautifulSoup


def get_data(url_visit):
    print("Đang truy cập",url_visit)
 
    html = requests.get(url_visit).text
    soup = BeautifulSoup(html,"html5lib")
    
    TieuDe = soup.find_all("h3",class_="post-title")
    DonGia = soup.find_all("span",class_="post-price")
    DienTich = soup.find_all("span",class_="post-acreage")
    DiaChi = soup.find_all("span",class_="post-location")
    NgayDang = soup.find_all("time",class_="post-time")
    
    str = ""
    for i in range (len(TieuDe)):
        ten = TieuDe[i].text.strip() if i <len(TieuDe) else ""
        gia = DonGia[i].text.strip() if i <len(TieuDe) else ""
        dientich = DienTich[i].text.strip() if i < len(TieuDe) else ""
        diachi = DiaChi[i].text.strip() if i < len(TieuDe) else ""
        ngaydang = NgayDang[i].text.strip() if i <len(TieuDe) else ""

        str += ten + "\n"
        str += gia + "\n"
        str += dientich + "\n"
        str += diachi + "\n"
        str += ngaydang + "\n"
    return str
def scrape_data(bđ,kt):
    data = ""
    for i in range(bđ,kt+1):
        url = f"https://phongtro123.com/tinh-thanh/da-nang?page={i}"
        page_data = get_data(url)
        data += page_data
    with open ("Phongtro.txt","w",encoding = "utf-8") as file: 
        file.write(data)
    print("Đã lưu dữ liệu vào file PhongTro")

scrape_data(1,30)
    
```

```python
#Câu 5. lấy link từ file RSS, bóc tách nội dung văn bản của trang ở dạng file.txt , tên file là tên đường link
import requests
import re
import os
from bs4 import BeautifulSoup
from urllib.parse import urlsplit

file_store="D:/Khai phá dữ liệu web/Luyện tập GK/noidung"

url_fileRSS="https://vnexpress.net/rss/suc-khoe.rss"
domain= urlsplit(url_fileRSS).netloc

def url_to_file_name(url):
    url = str(url).strip().replace(' ', '_')
    return re.sub(r'(?u)[^-\w.]', '', url) + ".txt"

def save_toTXT(url):
    print("Saving file,...")
    
    html = requests.get(url).text
    soup = BeautifulSoup(html, "html5lib")
    
    try:
        Title = soup.find("h1",class_="title-detail")    
        Desc = soup.find("p",class_="description")    
        Detail = soup.find("article",class_="fck_detail")
        Content= Title.text + "\n" + Desc.text + "\n" + Detail.text
        
        filename=os.path.join(file_store, url_to_file_name(url))    
        with open(filename, 'w',encoding='utf-8') as f:
            f.write(Content)
    finally:
        return
                
def visit(url):
    print("** Now visiting:", url)       
    save_toTXT(url)

def getURL_from_fileRSS(url_fileRSS):
    xml = requests.get(url_fileRSS).text
    links=re.findall('<a href="(.*?)"', xml)
    
    links_todo=[]    
    for link_url in links:
        if link_url is None:
            continue      
        
        if urlsplit(link_url).netloc != domain:
            continue
        
        if link_url in links_todo:
            continue
        
        links_todo.append(link_url)

    return links_todo

links_todo = getURL_from_fileRSS(url_fileRSS)
while links_todo:
    url_to_visit = links_todo.pop()
    visit(url_to_visit)
```

```python
#Câu 6. Tự động quét hyperlink trên website, lấy toàn bộ thẻ html lưu ở dạng file.html, tên file là tên đường link của trang web
import requests
import re
import os
from bs4 import BeautifulSoup
from urllib.parse import urljoin, urlsplit

links_seen=[]
links_todo = ["https://vnexpress.net/"]

domain= urlsplit(links_todo[0]).netloc
file_store="D:\Khai phá dữ liệu web\Luyện tập GK\Cau6"

def url_to_file_name(url):
    url = str(url).strip().replace(' ', '_')
    return re.sub(r'(?u)[^-\w.]', '', url) + ".html"

def download(url):
    print("Saving file,...")
    filename=os.path.join(file_store, url_to_file_name(url))    
    with open(filename, 'w',encoding='utf-8') as the_html:
            the_html.write(requests.get(url).text)
    
def visit(url):
    global links_todo
    
    links_seen.append(url)
    print("** Now visiting:", url)
    
    html = requests.get(url).text
    html_soup = BeautifulSoup(html, "html5lib")
    download(url)
    
    Count=0
    for link in html_soup.find_all("a"):
        link_url = link.get("href")  
        if link_url is None:
            continue
        
        full_url = urljoin(url, link_url)
        
        if urlsplit(full_url).netloc != domain:
            continue

        if urlsplit(full_url).query !="":
            continue
        
        if full_url in links_todo or full_url in links_seen:
            continue
        
        Count=Count+1
        links_todo.append(full_url)

    print(" + ",Count, "new link(s) found")
    print(" + ",len(links_seen), "link(s) seen/", len(links_todo), "link(s) need to do.")
    
    print("Enter to continues, ...")
    input()
```
