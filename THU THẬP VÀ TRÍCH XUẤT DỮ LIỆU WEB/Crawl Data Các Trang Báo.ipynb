```python
# Xoá stopword
import os
import glob
import re
from pathlib import Path

notebook_path = Path(os.getcwd()) / 'vietnamese-stopwords.txt'
# Đọc file stopwords
with open(notebook_path, "r", encoding="utf-8") as f:
    # Lấy danh sách stopwords
    List_StopWords = [word.strip() for word in f.read().split("\n") if word.strip()]

# Thư mục chứa các file cần xử lý
folder_path = 'D:/Crawl Data/NoiDung'
# Lấy danh sách tất cả các file .txt trong thư mục
all_files = glob.glob(os.path.join(folder_path, "*.txt"))

# Đọc và xử lý nội dung từ tất cả các file
for tên_file in all_files:
    with open(tên_file, 'r', encoding='utf-8') as file:
        dữ_liệu = file.read()
        
        # Loại bỏ số
        dữ_liệu = re.sub(r"\d+", "", dữ_liệu)
        
        # Loại bỏ stopwords
        dữ_liệu = " ".join(word for word in dữ_liệu.split() if word not in List_StopWords)

    # Lưu lại nội dung đã xử lý vào chính file
    with open(tên_file, 'w', encoding='utf-8') as file:
        file.write(dữ_liệu)

    print(f"Dữ liệu đã xử lý và lưu lại vào {tên_file}")
```

```python
import os
import glob
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer


folder_path = 'D:/Crawl Data/NoiDung'
all_files = glob.glob(os.path.join(folder_path, "*.txt"))


sentences = []
for tên_file in all_files:
    with open(tên_file, 'r', encoding='utf-8') as file:
        dữ_liệu = file.read()
        sentences.append(dữ_liệu)

# Tạo mô hình TF-IDF
tfidf_vectorizer = TfidfVectorizer()
tfidf_matrix = tfidf_vectorizer.fit_transform(sentences)

tfidf_df = pd.DataFrame(tfidf_matrix.toarray(), columns=tfidf_vectorizer.get_feature_names_out())

# Hiển thị các từ có TF-IDF cao nhất cho mỗi tài liệu
top_n = 3
top_keywords_per_doc = {}

for i, row in enumerate(tfidf_df.iterrows()):
    sorted_row = row[1].sort_values(ascending=False)  # Sắp xếp các từ trong tài liệu theo TF-IDF giảm dần
    top_keywords_per_doc[f"Tài liệu {i+1}"] = sorted_row.head(top_n).index.tolist()

# Hiển thị từ khóa TF-IDF cao nhất của từng tài liệu
print("Từ khóa có trọng số TF-IDF cao nhất trong mỗi tài liệu:")
for doc, keywords in top_keywords_per_doc.items():
    print(f"{doc}: {keywords}")

# Hiển thị các từ khóa có TF-IDF cao nhất trong toàn bộ tập dữ liệu
top_keywords_corpus = tfidf_df.max().sort_values(ascending=False).head(10)
print("\nTừ khóa có trọng số TF-IDF cao nhất trong toàn bộ tập dữ liệu:")
print(top_keywords_corpus)
```
